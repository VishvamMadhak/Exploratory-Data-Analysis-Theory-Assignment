**3)Reinforcement Learning**

```python
# VISHVAM MADHAK (21BDS0032)

import numpy as np
import random

# Grid
environment = np.array([
    [0, -1, 0, 0, 0],
    [0, 0, -1, -1, 0],
    [0, -1, 0, 0, 1],
    [0, 0, 0, -1, 0]
])

q_values = np.zeros((4, 5, 4))
moves = [(0, -1), (0, 1), (-1, 0), (1, 0)]

# Hyperparameters
alpha = 0.1
gamma = 0.9
exploration_rate = 0.2
iterations = 1000

# Check if a move is valid
def is_valid_move(row, col):
    if 0 <= row < environment.shape[0] and 0 <= col < environment.shape[1] and environment[row, col] != -1:
        return True
    return False

# Training the agent
for _ in range(iterations):
    row, col = 0, 0
    while (row, col) != (2, 4):
        if random.uniform(0, 1) < exploration_rate:
            action = random.choice(range(4))
        else:
            action = np.argmax(q_values[row, col])

        # Perform action
        delta_row, delta_col = moves[action]
        new_row, new_col = row + delta_row, col + delta_col

        if is_valid_move(new_row, new_col):
            # Calculate reward
            reward = 1 if (new_row, new_col) == (2, 4) else 0
            # Update Q-value
            q_values[row, col, action] = q_values[row, col, action] + alpha * (
                reward + gamma * np.max(q_values[new_row, new_col]) - q_values[row, col, action]
            )
            row, col = new_row, new_col

# Print the trained Q-table
print("Q-table after training:")
print(q_values)

# Derive the path from start to goal
row, col = 0, 0
trajectory = [(row, col)]
while (row, col) != (2, 4):
    action = np.argmax(q_values[row, col])
    delta_row, delta_col = moves[action]
    row, col = row + delta_row, col + delta_col
    trajectory.append((row, col))

print("Path to goal:", trajectory)
```
